```
courses = []
choice = 'Anything'

def line():
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

def show_menu():
    print("\33[1m\33[4mMENU" + "\33[")
    print('1. Add course')
    print('2. Unenroll course')
    print('3. View course')
    print('4. Exit')
    line()
    choice = input('Enter your choice:')
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        course = input('Enter your course name: ')
        courses.append(course)
        print('Course', course, 'added\n')
    elif choice == '2':
        course = input('Unenroll course: ')
        if course in courses:
            courses.remove(course)
            print('Unenroll the course', course)
        else:
            print('Could not find course', course)
    elif choice == '3':
        print('View the courses:')
        for course in courses:
            print(course)
    elif choice == '4':
        print('Thank you\n')
    else:
        print('Please enter one of 1 to 4\n')
```
